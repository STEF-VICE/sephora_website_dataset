```python
def histogram_boxplot(data, feature, figsize=(12, 7), kde=False, bins=None):
    plt.figure(figsize=figsize)
    plt.subplot(1, 2, 1)

    if bins is None:
        sns.histplot(data=data, x=feature, kde=kde)
    else:
        sns.histplot(data=data, x=feature, kde=kde, bins=bins)

    plt.xlabel(feature)
    plt.ylabel("Count")

    plt.subplot(1, 2, 2)
    sns.boxplot(data=data, y=feature)
    plt.ylabel(feature)

    plt.tight_layout()
    plt.show()
```

```python
def object_column_counts(data, column, n=10, figsize=(12, 6), fontsize=6):
    plt.figure(figsize=figsize)
    plt.title(f"Top {n} {column.capitalize()} Counts")
    column_counts = data[column].value_counts().head(n)
    sns.barplot(x=column_counts.index, y=column_counts.values)
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=70)
    plt.xticks(fontsize=fontsize)
    plt.show()
```

```python
def top_products_by_category(dt, category_column, product_column, brand_column, n_categories, n_products, figsize=(10, 6)):
    plt.figure(figsize=figsize)
    plt.title(f"Top {n_products} Products from Top {n_categories} Categories")

    top_categories = dt[category_column].value_counts().nlargest(n_categories).index.tolist()

    for category in top_categories:
        top_products = dt.loc[dt[category_column] == category, [product_column, brand_column]].value_counts().nlargest(n_products)
        product_labels = [f"{product} - {brand}" for product, brand in top_products.index]
        sns.barplot(x=top_products.values, y=product_labels, color='steelblue')
        plt.xlabel('Product Count')
        plt.ylabel('Product and Brand')
        plt.title(f"Top {n_products} Products in {category}")
        plt.show()
```

MODEL 1 - RANKED BASED RECOMMENDATION SYSTEM

```python
avg_rating = data.groupby('name').mean()['rating']

total_reviews = data.groupby('name').sum()['number_of_reviews']

brand = data.groupby('name')['brand'].first()

final_rating = pd.DataFrame({'name': avg_rating.index, 'avg_rating': avg_rating, 'total_reviews': total_reviews, 'brand': brand})
```

Obtaining the avarage rating and sum of the total reviews for every product in the data to recommend the top rated and reviewd products. Using this approach for new users, when entering the website they will see the top ranked products being recommended to them.

```python
final_rating.head()
```

```python
def top_n_products(data, n, min_reviews=None):
    if min_reviews is None:
        recommendations = data.sort_values(by='avg_rating', ascending=False)
    else:
        recommendations = data.sort_values(by=['avg_rating', 'total_reviews'], ascending=[False, False])
        recommendations = recommendations[recommendations['total_reviews'] > min_reviews]
    return recommendations.head(n)[['name', 'brand']]

new_user_recommendations = top_n_products(final_rating, n=5)
existing_user_recommendations = top_n_products(final_rating, n=5, min_reviews=100)

recommended_products_new_users = new_user_recommendations.apply(lambda row: f"{row['name']} ({row['brand']})", axis=1).tolist()
recommended_products_existing_users = existing_user_recommendations.apply(lambda row: f"{row['name']} ({row['brand']})", axis=1).tolist()

```

Model 2 -  COLLABORATIVE RECOMMENDATION SYSTEM (ITEM-ITEM)

```python
from sklearn.metrics.pairwise import cosine_similarity

selected_columns = ['brand', 'name', 'category', 'price']
data = data[selected_columns].copy()

data.dropna(inplace=True)

encoded_data = pd.get_dummies(data)

item_similarity = cosine_similarity(encoded_data)
```

I am using a cosine similarity metric, using cosine similarity in this case allows me to determine how similar two items are based on their brand, name, category, and price.

```python
def get_recommendations(brand, category, name, price_range=(0, 100), top_n=5):
    encoded_data_indices = encoded_data.index

    matching_items = data[(data['brand'] == brand) & (data['category'] == category) & (data['name'] == name)]

    if matching_items.empty:
        print("No matching product found.")
        return pd.DataFrame()

    item_index = matching_items.index[0]
    encoded_item_index = encoded_data_indices[item_index]

    item_scores = item_similarity[encoded_item_index]

    similar_items_indices = [i for i, score in enumerate(item_scores) if
                             (data.iloc[encoded_data_indices[i]]['category'] == category) and
                             (data.iloc[encoded_data_indices[i]]['brand'] != brand) and
                             (price_range[0] <= data.iloc[encoded_data_indices[i]]['price'] <= price_range[1])]

    similar_items_scores = item_scores[similar_items_indices]

    similar_items = data.iloc[encoded_data_indices[similar_items_indices]].copy()

    similar_items.loc[:, 'score'] = similar_items_scores

    similar_items = similar_items.sort_values(by='score', ascending=False)

    top_recommendations = similar_items.head(top_n)

    return top_recommendations[['brand', 'name', 'category', 'price']]
```

```python
brand = 'HUDA BEAUTY'
category = 'Eye Palettes'
name = 'Obsessions Eyeshadow Palette'
price_range = (0, 100)

recommendations = get_recommendations(brand, category, name, price_range)
print(recommendations)
```

Obtaining the similar item to the Huda Beauty Obsessions Eyeshadow Palette we get the above top 5 similar items based on Category and Price.

```python
brand = "NARS"
category = "Foundation"
name = "Sheer Glow Foundation"
price_range = (0, 30)

recommendations = get_recommendations(brand, category, name, price_range)
print(recommendations)
```

MODEL 3 - HYBRID RECOMMENDATION SYSTEM (COLLABORATIVE AND CONTENT BASED)

---



```python
def get_recommendations(brand, category, name, price_range=(0, 100), top_n=5, min_rating=4.5):
    item_index = df[(df['brand'] == brand) & (df['category'] == category) & (df['name'] == name)].index[0]
    item_scores = item_similarity[item_index]

    similar_items_indices = [i for i, score in enumerate(item_scores) if
                             (df.iloc[i]['category'] == category) and
                             (df.iloc[i]['brand'] != brand) and
                             (price_range[0] <= df.iloc[i]['price'] <= price_range[1]) and
                             (df.iloc[i]['rating'] >= min_rating)]

    similar_items_scores = item_scores[similar_items_indices]

    similar_items = df.iloc[similar_items_indices].copy()

    similar_items.loc[:, 'score'] = similar_items_scores

    similar_items = similar_items.sort_values(by='score', ascending=False)

    top_recommendations = similar_items.head(top_n)

    return top_recommendations[['brand', 'name', 'category', 'price']]

```

Using a hybrid model, we still use the cosine similarity to obtain the similar items to make recommendations but we have added a criterion of minimum rating of 4.5 to filter out item below that treshold.

```python
brand = 'HUDA BEAUTY'
category = 'Eye Palettes'
name = 'Obsessions Eyeshadow Palette'
top_n = 5
min_rating = 4.5

recommendations = get_recommendations(brand, category, name, price_range, top_n, min_rating)

print(recommendations)
```

Obtaining the top 5 products similar to the Huda Beauty Obsessions Eyeshadow Palette with a minimum rating of 4.5
